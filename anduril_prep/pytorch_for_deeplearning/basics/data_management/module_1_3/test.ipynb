```python
import os
import zipfile
from datetime import datetime

# Create a timestamped zip filename to avoid overwriting
timestamp = datetime.now().strftime('%Y%m%d_%H%M%S')
zip_filename = f'archive_{timestamp}.zip'

print(f'Creating zip file: {zip_filename}')

# Get the current directory
current_dir = os.getcwd()

# Create the zip file
with zipfile.ZipFile(zip_filename, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for root, dirs, files in os.walk(current_dir):
        for file in files:
            file_path = os.path.join(root, file)
            # Use relative path inside the zip to preserve folder structure
            arcname = os.path.relpath(file_path, current_dir)
            zipf.write(file_path, arcname)
            print(f'Added: {arcname}')

print(f'\nZip creation complete!')
print(f'File saved as: {zip_filename}')
print(f'Size: {os.path.getsize(zip_filename) / (1024 * 1024):.2f} MB')
```
